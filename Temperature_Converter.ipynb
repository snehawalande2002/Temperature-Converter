```python
class Temperature:
    temp = 0

    def __init__(self, temp):
        self.temp = temp
    def farh(self):
        result = float((9 * self.temp) / 5 + 32)
        return result

    def cel(self):
        result = float((self.temp - 32) * 5 / 9)
        return result


input1 = float(input("Input temperature in celsius: "))
temp1 = Temperature(input1)
print(temp1.farh())

input1 = float(input("Input temperature in fahrenheit: "))
temp1 = Temperature(input1)
print(temp1.cel())
```
